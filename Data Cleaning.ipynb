```python
import pandas as pd
from IPython.display import display, Markdown
```

```python
display(Markdown("## 📂 1. Load & Clean Data"))

# Load the dataset
df = pd.read_csv(r"C:\Users\bless\Downloads\online_retail.csv", encoding='latin1')

# Display BEFORE cleaning
display(Markdown("### 🔴 BEFORE CLEANING"))
print(f"📊 Raw Data Shape: {df.shape[0]:,} rows × {df.shape[1]} columns")

# Check for issues
print(f"\n🔍 Data Issues Detected:")
print(f"   - Missing CustomerID: {df['CustomerID'].isnull().sum():,}")
print(f"   - Missing Description: {df['Description'].isnull().sum():,}")
print(f"   - Cancelled orders (C prefix): {df[df['InvoiceNo'].astype(str).str.startswith('C', na=False)].shape[0]:,}")
print(f"   - Negative Quantity: {(df['Quantity'] < 0).sum():,}")
print(f"   - Negative/Zero Price: {(df['UnitPrice'] <= 0).sum():,}")

# CLEANING PROCESS
print("\n🧹 Performing Data Cleaning...")

# Remove missing values
df_clean = df.dropna(subset=['CustomerID', 'Description'])

# Remove cancelled transactions
df_clean = df_clean[~df_clean['InvoiceNo'].astype(str).str.startswith('C')]

# Remove negative/zero quantities
df_clean = df_clean[df_clean['Quantity'] > 0]

# Remove negative/zero prices
df_clean = df_clean[df_clean['UnitPrice'] > 0]

# Convert date and create revenue
df_clean['InvoiceDate'] = pd.to_datetime(df_clean['InvoiceDate'])
df_clean['Revenue'] = df_clean['Quantity'] * df_clean['UnitPrice']

# Display AFTER cleaning
display(Markdown("### 🟢 AFTER CLEANING"))
print(f"📊 Cleaned Data Shape: {df_clean.shape[0]:,} rows × {df_clean.shape[1]} columns")
print(f"📅 Date Range: {df_clean['InvoiceDate'].min().date()} → {df_clean['InvoiceDate'].max().date()}")
print(f"👥 Unique Customers: {df_clean['CustomerID'].nunique():,}")
print(f"📦 Unique Products: {df_clean['StockCode'].nunique():,}")
print(f"💰 Total Revenue: {df_clean['Revenue'].sum():,.2f}")

# Sample data
display(Markdown("### 📋 Sample Data (First 5 Rows)"))
display(df_clean.head())

# Save cleaned dataset 
df_clean.to_csv('cleaned_online_retail.csv', index=False)
```
